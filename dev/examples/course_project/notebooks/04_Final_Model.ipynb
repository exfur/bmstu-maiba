## 📥 Шаг 1. Инициализация окружения и загрузка тензоров

Импортируем алгоритмы классификации и метрики качества из `scikit-learn`. Загрузим обучающую и тестовую выборки, сохраненные на предыдущем шаге.


```python
import os
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import warnings

warnings.filterwarnings('ignore')

PROCESSED_DIR = os.path.join(".", "data", "processed")

# Загрузка подготовленных выборок
try:
    X_train = pd.read_csv(os.path.join(PROCESSED_DIR, "X_train.csv"))
    X_test = pd.read_csv(os.path.join(PROCESSED_DIR, "X_test.csv"))
    y_train = pd.read_csv(os.path.join(PROCESSED_DIR, "y_train.csv")).squeeze() # squeeze для перевода в Series
    y_test = pd.read_csv(os.path.join(PROCESSED_DIR, "y_test.csv")).squeeze()
    print(f"✅ Данные успешно загружены! Размер трейна: {X_train.shape}, Размер теста: {X_test.shape}")
except FileNotFoundError:
    print("⚠️ Файлы выборок не найдены. Вернитесь к ноутбуку 02_Preprocessing и выполните Задание 4.")

# Словарь для хранения результатов всех моделей
model_results = {}
```

## 🛠 ЗАДАНИЕ 1: Обучение логистической регрессии (Logistic Regression)
**Бизнес-контекст:** Логистическая регрессия — это индустриальный стандарт "первого приближения". Это быстро обучаемая, легко интерпретируемая линейная модель, которая покажет базовый уровень (baseline), от которого мы будем отталкиваться при оценке более сложных ансамблей.

**Инструкция (TODO):**
1. Инициализируйте объект `LogisticRegression` с параметром `random_state=42`.
2. Обучите модель на тренировочных данных (`X_train`, `y_train`).
3. Сделайте предсказания классов и их вероятностей на тестовой выборке (`X_test`).
4. Рассчитайте метрики качества (ROC-AUC, F1, Precision, Recall).

*🤖 Помощь AI-тьютора:*
* `#ML_TASK1_START` — для расчета вероятностей (нужны для ROC-AUC) используйте метод `.predict_proba(X_test)[:, 1]`.


```python
# [MASTER SOLUTION]
if 'X_train' in locals():
    # 1. Инициализация и обучение
    log_reg = LogisticRegression(random_state=42)
    log_reg.fit(X_train, y_train)
    
    # 2. Предсказания
    y_pred_lr = log_reg.predict(X_test)
    y_proba_lr = log_reg.predict_proba(X_test)[:, 1]
    
    # 3. Оценка качества
    metrics_lr = {
        "ROC-AUC": roc_auc_score(y_test, y_proba_lr),
        "F1-Score": f1_score(y_test, y_pred_lr),
        "Precision": precision_score(y_test, y_pred_lr),
        "Recall": recall_score(y_test, y_pred_lr)
    }
    
    model_results["Logistic Regression"] = metrics_lr
    print("✅ Логистическая регрессия обучена. Метрики:")
    for m, v in metrics_lr.items(): print(f"  - {m}: {v:.4f}")
```

## 🛠 ЗАДАНИЕ 2: Обучение случайного леса (Random Forest)
**Бизнес-контекст:** Клиентское поведение часто описывается нелинейными закономерностями (например, клиенты с высоким доходом уходят по иным причинам, нежели клиенты с низким). `RandomForest` строит сотни независимых деревьев принятия решений, что позволяет выявлять скрытые нелинейные связи и снижает риск переобучения.

**Инструкция (TODO):**
1. Инициализируйте `RandomForestClassifier` (ограничьте глубину `max_depth=10`, чтобы деревья обучались быстрее, `random_state=42`).
2. Выполните цикл обучения и прогнозирования.
3. Сохраните метрики в общий словарь `model_results`.

*🤖 Помощь AI-тьютора:*
* `#ML_TASK2_START` — синтаксис обучения идентичен логистической регрессии благодаря единому API scikit-learn.


```python
# [MASTER SOLUTION]
if 'X_train' in locals():
    # 1. Инициализация и обучение
    rf_model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
    rf_model.fit(X_train, y_train)
    
    # 2. Предсказания
    y_pred_rf = rf_model.predict(X_test)
    y_proba_rf = rf_model.predict_proba(X_test)[:, 1]
    
    # 3. Оценка качества
    metrics_rf = {
        "ROC-AUC": roc_auc_score(y_test, y_proba_rf),
        "F1-Score": f1_score(y_test, y_pred_rf),
        "Precision": precision_score(y_test, y_pred_rf),
        "Recall": recall_score(y_test, y_pred_rf)
    }
    
    model_results["Random Forest"] = metrics_rf
    print("✅ Случайный лес обучен. Метрики:")
    for m, v in metrics_rf.items(): print(f"  - {m}: {v:.4f}")
```

## 🛠 ЗАДАНИЕ 3: Градиентный бустинг (Gradient Boosting)
**Бизнес-контекст:** Градиентный бустинг (Gradient Boosting) — это один из самых мощных алгоритмов для табличных данных. В отличие от леса, где деревья независимы, в бустинге каждое следующее дерево целенаправленно исправляет ошибки предыдущих.

**Инструкция (TODO):**
1. Инициализируйте `GradientBoostingClassifier` с параметром `random_state=42`.
2. Обучите модель и получите предсказания.
3. Добавьте результаты в `model_results`.

*🤖 Помощь AI-тьютора:*
* `#ML_TASK3_START` — базовые параметры GBC часто уже дают отличный результат из коробки.


```python
# [MASTER SOLUTION]
if 'X_train' in locals():
    # 1. Инициализация и обучение
    gbc_model = GradientBoostingClassifier(random_state=42)
    gbc_model.fit(X_train, y_train)
    
    # 2. Предсказания
    y_pred_gbc = gbc_model.predict(X_test)
    y_proba_gbc = gbc_model.predict_proba(X_test)[:, 1]
    
    # 3. Оценка качества
    metrics_gbc = {
        "ROC-AUC": roc_auc_score(y_test, y_proba_gbc),
        "F1-Score": f1_score(y_test, y_pred_gbc),
        "Precision": precision_score(y_test, y_pred_gbc),
        "Recall": recall_score(y_test, y_pred_gbc)
    }
    
    model_results["Gradient Boosting"] = metrics_gbc
    print("✅ Градиентный бустинг обучен. Метрики:")
    for m, v in metrics_gbc.items(): print(f"  - {m}: {v:.4f}")
```
